### FUNÇÃO PARA SEGUIDORES E AMIGOS

```python
#Função para mostrar o número de seguidores de determinado perfil
def user_followers(perfil):
  user = api.get_user(screen_name=perfil)
  print('Usuário desejado: ',user.screen_name)
  print('Total de seguidores: ',user.followers_count,'\n')
  print(f'Amigos de {perfil} no Twitter:')
  for friend in user.friends():
    print(friend.screen_name)
```

```python
user_followers('BarackObama')
```

### FUNÇÃO PARA TWEETS COM DETERMINADA PALAVRA-CHAVE

```python
last_month= date.today() - timedelta(days=30)
```

```python
#Função para buscar posts e tendências de acordo com determinada palavra-chave
def keyword(palavra, quantidade):
  for tweet in tweepy.Cursor(api.search,
                            q=palavra,
                            since=last_month.strftime('%Y-%m-%d'),
                            until=date.today().strftime('%Y-%m-%d'),
                            include_entities=False,
                            lang='pt').items():
    print(tweet.text)
    if len(tweet.text) == quantidade:
      break
```
